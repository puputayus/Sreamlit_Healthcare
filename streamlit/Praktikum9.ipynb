```python
import pandas as pd
from pycaret.datasets import get_data
from pycaret.classification import *

data = pd.read_csv('C:/Users/asuss/Downloads/streamlit/water_potability.csv')
data.fillna(data.mean(), inplace=True)
s = setup(data, target='Potability', session_id=123)
best = compare_models()
model = tune_model(best)
evaluate_model(model)
finalize_model(model)
save_model(model, 'water_pipeline')
```

```python
import pandas as pd
from pycaret.datasets import get_data
from pycaret.clustering import *

# Load dataset
data = pd.read_csv('C:/Users/asuss/Downloads/streamlit/water_potability.csv')
data.fillna(data.mean(), inplace=True)

# Setup clustering
s = setup(data, session_id=123)

# Create clustering model
kmeans = create_model('kmeans')

# Assign model
assigned_model = assign_model(kmeans)

# Evaluate the model
evaluate_model(assigned_model)

# Save the model
save_model(assigned_model, 'clustering_pipeline')
```
